### Import usefull libraries :

```python
import keras_ocr
import matplotlib.pyplot as plt
%matplotlib inline
```

### Defining keras_ocr pipeline and initializing it to 'pipeline' variale :

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

### Creating a list of images for Text Recognition : 

```python
images = [
    keras_ocr.tools.read(images) for images in [
        'https://i.pinimg.com/originals/90/e0/b1/90e0b16039dacc8c318c5bf1642495c4.png',
        'https://didyouknowcars.com/wp-content/uploads/2020/07/a-road-sign-along-a-desert-highway-780x470.jpg',
        'https://www.lostpaths.com/wp-content/uploads/2018/03/2018-03-07-00-CA-Death-Valley-services-sign-e1520460343694.jpg',
        'https://i.imgur.com/I6508bVl.jpg'         
                                                ]
        ]
```

# Recognizing/Predicting Text from images :

```python
prediction = pipeline.recognize(images)
```

```python
prediction
```

# Plotting recognized text and images together :

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(20, 20))
for ax, image, predictions in zip(axs, images, prediction):
    keras_ocr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
